```python
import geopandas as gpd
import matplotlib.pyplot as plt
import osmnx as ox
import networkx as nx
import pandas as pd
import geopandas as gpd
import pickle
import numpy as np

calgary_graph = ox.io.load_graphml('./mapGraphs/calgary_drive_network_updated2.graphml')

# Convert the graph to GeoDataFrame with geometry
edges_gdf, nodes_gdf = ox.graph_to_gdfs(calgary_graph)

# Plot the joined data
ax = nodes_gdf.plot(column="edge_tree_count", linewidth=0.5, figsize=(10, 10))

# Set the axis labels and title
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title("Calgary Road Tree Count")

# Show the plot
plt.show()
```
